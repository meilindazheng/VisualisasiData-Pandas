### [25.3] Pengaturan Tampilan
- Pengaturan max rows digunakan untuk menampilkan maksimal jumlah baris yang ditampilkan secara default yaitu 5
- Pengaturan max column digunakan untuk menampilkan maksimal jumlah kolom yang ditampilkan secara default yaitu 6
- Pengaturan max colwidth digunakan untuk menampilkan lebar kolom yang ditampilkan secara default yaitu 20 karakter

```python
pd.set_option('display.max_rows',5)
pd.set_option('display.max_columns',6)
pd.set_option('display.max_colwidth',20)

df
```

```python
pd.reset_option('^display.',silent=True)
df
```
