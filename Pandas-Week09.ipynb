## [33] Menampilkan Nilai Cummulative (Kumulatif)

### [33.1] Import Module

```python
import pandas as pd
print (pd.__version__)
```

### [33.2] Persiapan Data Frame
- Data frame yang dibentuk terdiri dari dua kolom yaitu pemain dan goal
- Data pada kolom gol mengindikasikan jumlah gol yang berhasil dicetak oleh pemain dalam satu sesi pertandingan

```python
d = {'pemain':['Budi','Joni','Iwan','Budi','Budi','Iwan','Asep',"Joni"],
    'goal':[2,1,3,1,1,2,2,3]}
df = pd.DataFrame(d)
df
```

### [33.3] Menampilkan Nilai Kumulatif
- Fungsi cumsum() digunakan untuk menampilkan data secara kumulatif, dalam hal ini data pada kolom 'goal' akan ditambahkan dari data di kolom sebelumnya.
- Fungsi cumsum() yang diikutkan setelah penggunaan fungsi groupby() berguna untuk menampilkan nilai kumulatif data yang dikelompokan berdasarkan pemain.
- Fungsi cummax() digunakan untuk menampilkan nilai tertinggi sebuah data sampai dengan sebuah kolom tertentu
- Fungsi cummin() digunakan untuk menampilkan nilai terendah sebuah data sampai dengan sebuah kolom tertentu
- Fungsi cumprod() digunakan untuk menampilkan hasil perkalian data kumulatif sebuah data sampai dengan sebuah kolom tertentu

```python
df['goal'].cumsum().to_frame()
```

```python
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()
df
```

```python
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()
df
```

```python
df['cummax'] = df['goal'].cummax()
df
```

```python
df['cummin'] = df['goal'].cummin()
df
```

```python
df['cumprod'] = df['goal'].cumprod()
df
```

## [34] Mapping pada Data Frame dengan applymap()

### [34.1] Import Module

```python
import pandas as pd
print(pd.__version__)
```

### [34.2] Persiapan Data Frame
- Data frame yang dibentuk terdiri dari 3 kolom yaitu jenis kelamin, usia, shift

```python
df = pd.DataFrame({'jenis_kelamin':['Pria','Wanita','lelaki','Lelaki','perempuan'],'usia':[23,21,24,22,21],'shift':['pagi','siang','Malam','Siang','pagi']})
df
```

### [34.3] Mapping Data dengan menggunakan applymap()
- Fungsi dibawah akan menguji tipe data dari sebuah kolom, jika tipe data merupakan string maka string akan dijadikan lower,sedangkan jika tipe data bukan str maka data akan dikembalikan apa adanya

```python
df = df.applymap(lambda x: x.lower() if type(x)==str else x)
df
```

### [34.4] Applymap() dengan dictionary
- Applymap() dengan dictionary di bawah ini digunakan untuk menstandarisasi lambang dari data pada kolom sebagai contoh pria dan laki-laki akan digantikan dengan nilai 'L'

```python
mapping = {'pria':'L',
          'lelaki':'L',
          'wanita':'P',
          'perempuan':'P',
          'pagi':1,
          'siang':2,
          'malam':3}
df.applymap(mapping.get)
```

```python
df[['jenis_kelamin','shift']] = df[['jenis_kelamin','shift']].applymap(mapping.get)
df
```

## [35] Memadukan fungsi agregasi dengan transform()

### [35.1] Import Module

```python
import pandas as pd
print(pd.__version__)
```

### [35.2] Persiapan Data Frame
- Data Frame di bawah berisi 3 kolom yaitu 'no_nota' ,'kopi','harga'
- Data sebagai isi dari kolom pada dataframe dibentuk dalam list

```python
d = {'no_nota':[1,1,1,2,2,3,4,5],
    'kopi':['latte','cappucino','espresso','latte','espresso','cappucino','latte','espresso'],
    'harga':[50,60,80,150,120,60,100,40]}
df = pd.DataFrame(d)
df
```

### [35.3] Menghitung total harga untuk setiap nota
- Fungsi groupby() digunakan untuk mengelompokan data pada dataframe berdasarkan sebuah kategori tertentu
- Fungsi sum digunakan untuk menghitung total dari nilai data

```python
df.groupby('no_nota')['harga'].sum().to_frame()
```

```python
df['total_harga'] = df.groupby('no_nota')['harga'].transform(sum)
df
```

### [35.4] Menghitung total omset untuk setiap jenis kopi yang terjual

```python
df.groupby('no_nota')['harga'].sum().to_frame()
```

```python
df['total_omset'] = df.groupby('kopi')['harga'].transform(sum)
df
```

## [36] Menyatukan kolom dengan str.cat()

### [36.1] Import Module

```python
import pandas as pd
print(pd.__version__)
```

### [36.2] Persiapan Data Frame

```python
data = {'nama':['bayu','indra','devi','agni'],
       'jenis_kelamin':['L','L','P','L'],
       'usia':[23,21,22,25]}
df = pd.DataFrame(data)
df
```

### [36.3] Menyatukan data dengan str.cat()
- JIka tipe data pada kolom berbeda maka bisa dikonversikan menjadi string dengan method astype()

```python
df['nama'].str.cat(df['jenis_kelamin'],sep=',').to_frame()
```

```python
df['nama_jk'] = df['nama'].str.cat(df['jenis_kelamin'],sep=',')
df
```

```python
df['nama'].str.cat(df['usia'].astype('str'),sep='-').to_frame()
```

```python
df['nama_usia'] = df['nama'].str.cat(df['usia'].astype('str'),sep='-').to_frame()
df
```
