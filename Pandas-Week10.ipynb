## [38] Mengenal parameter Header dan skiprows

### [38.1] Import Module
- Penggunaan parameter header adalah untuk menentukan baris mana yang akan dijadikan header dari sebuah dataframe
- Sedangkan untuk parameter skiprows adalah untuk melompati atau mengabaikan data pada kolom tertentu

```python
import pandas as pd
print(pd.__version__)
```

```python
df = pd.read_csv('./iris_error.csv')
df.head(8)
```

```python
df = pd.read_csv('./iris_error.csv',header=2,skiprows=[5,6])
df.head()
```
